```python
# Percentage of NaN values in data frame
def PercentNaN(df,feature):
    f_NaN = df[feature].isnull().sum()*100.0/df.shape[0]
    print "% of NaN values for ",feature,"{0:.2f}".format(f_NaN)
    return f_NaN

bonus_NaN = PercentNaN(df,'bonus')
total_payments_NaN = PercentNaN(df,'total_payments')
total_stock_value_NaN = PercentNaN(df,'total_stock_value')
```

```python
# Plot Missing Values Graph
plt.style.use('seaborn-white')

labels = ['bonus NaN', 'total_payments NaN', 'total_stock_value NaN']
x_pos = np.arange(len(labels))
width = 0.8
vals = (bonus_NaN,total_payments_NaN,total_stock_value_NaN)

plt.bar(x_pos,vals)
plt.suptitle('% of Missing Values Among Features')
plt.xticks(x_pos)
plt.xticks(x_pos+width/2.0, labels)
plt.ylabel('Percentage')
axes = plt.gca()
axes.set_ylim([0,100])
# Save figure in file
plt.savefig('MissingDataAmongFeatures.png')
plt.show()
```

```python
# Data Quality Analysis function
# i.e. Is there any missing and duplicate value for given feature and given data frame? 
def dataqualityanalysis(df,feature):
    print "# unique values: ",len(df[feature].unique())
    print "# null values:",df[feature].isnull().sum()
    if len(df.set_index(feature).index.get_duplicates()) > 0:
        print "# duplicates: ", len(df.set_index(feature).index.get_duplicates())
        print "duplicates are:\n"
        print df.set_index(feature).index.get_duplicates()
```

## Process empty values

```python
enron_data = df.fillna(0)
```

```python
enron_data.describe(include='all')
```

## Model: k-means

```python
X = enron_data[['total_payments','total_stock_value']].as_matrix()
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
enron_data.columns.values
```

```python
# Overall view
plt.plot(X[:,0], X[:,1],'.',color='red',alpha=0.2)
plt.title("total_payments vs total_stock_value")
plt.xlabel("total payments")
plt.ylabel("total stock value")
#plt.xlim(-1e8,1e8)
#plt.ylim(-1e7,1e7)
plt.show()
```
